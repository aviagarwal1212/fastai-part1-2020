### Building a bigger decision tree

```python
m = DecisionTreeRegressor()
m.fit(train_xs, train_y)
```

Creating a function to measure the root mean squared error of the model

```python
def r_rmse(pred, y): return round(math.sqrt(((pred-y)**2).mean()), 6)
def model_rmse(m, xs, y): return r_rmse(m.predict(xs),y)
```

```python
model_rmse(m, train_xs, train_y)
```

```python
model_rmse(m,valid_xs, valid_y)
```

Looks like we are overfitting

```python
m.get_n_leaves(), len(train_xs)
```

Stopping such that each node has atleast 25 records

```python
m = DecisionTreeRegressor(min_samples_leaf=25)
m.fit(train_xs, train_y);
```

```python
model_rmse(m, train_xs, train_y), model_rmse(m,valid_xs, valid_y)
```

```python
m.get_n_leaves()
```

### Building a Random Forest

```python
def rf(xs, y, n_estimators=40, max_samples=200_000, max_features=0.5, min_samples_leaf=5, **kwargs):
    return RandomForestRegressor(
        n_jobs = -1,
        n_estimators = n_estimators,
        max_samples = max_samples,
        max_features = max_features,
        min_samples_leaf = min_samples_leaf,
        oob_score = True
    ).fit(xs, y)
```

```python
m = rf(train_xs, train_y)
```

```python
model_rmse(m, train_xs, train_y), model_rmse(m, valid_xs, valid_y)
```

Validating that its taking the mean of the decision trees

```python
preds = np.stack([t.predict(valid_xs) for t in m.estimators_])
r_rmse(preds.mean(0),valid_y)
```

How RMSE improves as we add more trees to a random forest

```python
plt.plot([r_rmse(preds[:i+1].mean(0), valid_y) for i in range(40)]);
```

#### Feature Importance

```python
def rf_importance(m, df):
    return pd.DataFrame(
        {'cols': df.columns, 'imp':m.feature_importances_},
    ).sort_values('imp', ascending=False)
```

```python
def plot_fi(fi):
    return fi.plot('cols', 'imp', 'barh', figsize=(12,7), legend=False)

plot_fi(fi[:30])
```

### Neural Network

```python
df_nn = pd.read_csv(path/'TrainAndValid.csv', low_memory=False)
```

Treating ordinal columns:

```python
sizes = 'Large', 'Large / Medium', 'Medium', 'Small', 'Mini', 'Compact'
df_nn['ProductSize'] = df_nn['ProductSize'].astype('category')
df_nn['ProductSize'].cat.set_categories(sizes, ordered=True, inplace=True)
```

Defining dependant variables:

```python
dep_var = 'SalePrice'
df_nn[dep_var] = np.log(df_nn[dep_var])
```

Adding date-related features:

```python
df_nn = add_datepart(df_nn, 'saledate')
```

Limiting to features we have gathered are important per random forest:

```python
df_nn_final = df_nn[list(train_xs_time.columns) + [dep_var]]
```

Splitting continuous and categorical features:

```python
cont_nn, cat_nn = cont_cat_split(df_nn_final, max_card=9000, dep_var=dep_var)
```

Creating `TabularPandas`

```python
procs_nn = [Categorify, FillMissing, Normalize]
to_nn = TabularPandas(df_nn_final, procs_nn, cat_nn, cont_nn,
                      splits=splits, y_names=dep_var)
```

Creating dataloaders

```python
dls = to_nn.dataloaders(bs=1024)
```

Since its a regression problem, we need to figure out out `y_range`

```python
y = dls.train.y
```

```python
y.min(), y.max()
```

Defining learner

```python
learn = tabular_learner(dls, layers=[500,250], n_out=1, y_range=(8,12), loss_func=F.mse_loss)
```

Training

```python
learn.lr_find()
```

```python
learn.fit_one_cycle(5, 1e-2)
```

Getting `r_rmse` loss

```python
predictions, targets = learn.get_preds()
r_rmse(predictions, targets)
```

Ensembling with random forest

```python
rf_predict = m1.predict(valid_xs_time)
nn_predict = predictions
rf_predict.shape, nn_predict.shape
```

```python
nn_predict = to_np(predictions.squeeze())
ens_predict = (rf_predict + nn_predict)/2
```

```python
r_rmse(ens_predict, valid_y)
```
